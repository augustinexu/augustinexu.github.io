```python
import pandas as pd
import bqplot
import numpy as np
import traitlets
import ipywidgets
import matplotlib.pyplot as plt
```

```python
buildings = pd.read_csv("building_inventory.csv",
                       na_values={'Year Acquired': 0})
buildings = buildings.dropna(subset=['Year Acquired'])
```

```python
buildings
```

```python
data = buildings[['Agency Name', 'Congress Dist', 'Square Footage']]
data = data.groupby(['Agency Name','Congress Dist'])['Square Footage'].sum().reset_index(name='Total Square Footage')
lst = data['Agency Name'].unique().tolist()
data.sort_values(by='Agency Name')
'''def set_index(x):
    if x in lst:
        return lst.index(x)'''
```

```python
row = sorted(data['Congress Dist'].unique())
col = data['Agency Name'].unique()
new = pd.DataFrame(columns= col, index = row )
for index in data.index:
    new.loc[data['Congress Dist'][index],data['Agency Name'][index]] = data['Total Square Footage'][index]
new = new.fillna(0)
new
```

```python
i,j = 13,33
right = buildings[(buildings['Agency Name'] == col[j]) & (buildings['Congress Dist'] == i)]
right = right.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Square Footage')
x_sch = bqplot.LinearScale()
y_sch = bqplot.LinearScale()
x_axh = bqplot.Axis(scale=x_sch, label='Year')
y_axh = bqplot.Axis(scale=y_sch, label='Total Footage Acquried', orientation='vertical')
line = bqplot.Lines(x = right['Year Acquired'], y = right['Total Square Footage'],scales={'x': x_sch, 'y': y_sch})
figh = bqplot.Figure(marks=[line], axes=[x_axh, y_axh])
```

```python
mySelectedLabel = ipywidgets.Label()
x_sc = bqplot.OrdinalScale() # categorical data
y_sc = bqplot.OrdinalScale()
col_sc = bqplot.ColorScale(scheme='Reds')

# 3. Axes
ax_x = bqplot.Axis(scale=x_sc)
ax_y = bqplot.Axis(scale=y_sc, orientation='vertical')
ax_col = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right') # colorbar

# 4. Mark (gridheat map)
heat_map = bqplot.GridHeatMap(color = new,
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'}, # on clicking on a grid, select that grid
                             anchor_style={'fill':'blue'}, # turn the grid blue on selection
                             selected_style={'opacity':1.0}, # selected grid is fully opaque on selection,
                             unselected_style={'opacity':0.8}) # otherwise, a little see-through

def on_selected(change):
    if len(change['owner'].selected) == 1: # forcing only selection of 1 grid
        i,j = change['owner'].selected[0] # only 1 selection
        v = new.iloc[i,j] # value of mean of the data at the i,j grid
        mySelectedLabel.value = str(col[j]) + ' in Congress Dist ' + str(i) + ' Total Square Footage =' + str(v)
        right = buildings[(buildings['Agency Name'] == col[j]) & (buildings['Congress Dist'] == i)]
        right = right.groupby(['Year Acquired'])['Square Footage'].sum().reset_index(name='Total Square Footage')
        line.x = right['Year Acquired']
        line.y = right['Total Square Footage']
        # don't forget the actual linking!
heat_map.observe(on_selected,'selected')

fig = bqplot.Figure(marks=[heat_map], axes=[ax_col, ax_x, ax_y])
# 5. Interactions (also in step 4 as well)
fig.layout.min_width='700px'
figh.layout.min_width='700px'
figures = ipywidgets.HBox([fig,figh])
myDashboard = ipywidgets.VBox([mySelectedLabel, figures])
myDashboard
```

1. In the building_inventory_csv, we generated one sub dataframe which includes the columns 'Agency name', 'Congress Dist' and 'Square Footage'. Then the dataframe was groubyed by 'Agency name' 'Congress Dist' and sum the 'Square footage' into 'Total Square Footage' and sorted the 'Agency name'. Finally, we generated an empty 19 by 35 dataframe which 19 rows are the 19 congress districts and 35 columns are the 35 agencies. We inserted the values into the new empty dataframe and fill 0s for those agencies which don't have buildings in some districts. 

2. For the NaNs (empty entries), I decided to replce the all the 0s in 'Year acquried' with NaN and drop the rows only by 'Year acquried' since there is Congress Dist 0. The dropna function would drop these by default. The reason why only drop the 0s in 'Year acquried' is that the 0s in 'Year acquried' don't make any sense. However, the 0s in 'Square Footage' have no effects since we can calculate the total square footage. 

3. I didn't have any special choice for the colors while only use the default colors for in the sample code for grip heat map and default color for the line plot. The layout is the same as in class coding. I only changed the size since the data is larger and the plots need to show more information. The size needs to be larger in order to make the plot clear to see and read.
